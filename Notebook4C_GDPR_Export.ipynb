# 📘 Notebook 4C — GDPR-Compliant Dataset Export (Final Step)
This notebook merges original OCR text and anonymised text into a final GDPR-compliant dataset.
It loads outputs from 4A and 4B and produces the final export file **GDPR_Compliant_EN_ES.pkl**.

```python
import pandas as pdimport numpy as np
```

## Step 1 — Load Anonymised Dataset (from Notebook 4A)

```python
df_clean = pd.read_pickle('gdpr_clean_en_es.pkl')df_clean.head()
```

## Step 2 — Load PII Validation Dataset (from Notebook 4B)

```python
df_val = pd.read_pickle('gdpr_pii_validation_en_es.pkl')df_val.head()
```

## Step 3 — Merge Validation Columns

```python
df_clean['pii_before'] = df_val['pii_before']df_clean['pii_after'] = df_val['pii_after']df_clean.head()
```

## Step 4 — Export GDPR-Compliant Dataset

```python
df_export = df_clean[['ocr_text','ocr_text_clean','embedding','label','language','pii_before','pii_after']]df_export.to_pickle('GDPR_Compliant_EN_ES.pkl')print('Export complete → GDPR_Compliant_EN_ES.pkl')
```

## Step 5 — Final GDPR Audit Summary

```python
print('Total rows:', len(df_export))print('PII BEFORE anonymisation:', df_export['pii_before'].sum())print('PII AFTER anonymisation:', df_export['pii_after'].sum())print('GDPR-compliant dataset is ready.')
```
