# 📘 Notebook 4A — GDPR Anonymisation (JSON‑Escaped GPT‑4o Method)
This notebook performs GDPR anonymisation on multilingual OCR text (English + Spanish).
It uses a **JSON‑escaped text pipeline** to avoid breaking notebook structure.

**This notebook performs:**
- JSON‑escaped anonymisation prompts (safe for .ipynb)
- GPT‑4o‑mini anonymisation
- Saves anonymised text as `ocr_text_clean`
- Outputs `gdpr_clean_en_es.pkl`

```python
from openai import OpenAIimport pandas as pdimport numpy as npimport json, timefrom tqdm import tqdmclient = OpenAI(api_key='YOUR_OPENAI_KEY')
```

## Step 1 — Load Multilingual Dataset (from Notebook 2)

```python
df = pd.read_pickle('MultiFinBen_EN_ES_ready.pkl')df.head()
```

## Step 2 — GDPR Anonymisation Function (JSON‑escaped text)

```python
def anonymize_text(text):    if text is None or len(text.strip()) == 0:        return text    # Escape text for safe notebook insertion    safe_text = json.dumps(text)[1:-1]    prompt = (        'Remove all personal data such as names, emails, phone numbers, addresses, ' +        'and ID numbers. Replace with generic placeholders. Preserve financial meaning. ' +        'Text: ' + safe_text    )    response = client.chat.completions.create(        model='gpt-4o-mini',        messages=[{'role':'user','content': prompt}]    )    return response.choices[0].message.content
```

## Step 3 — Apply Anonymisation

```python
df['ocr_text_clean'] = [anonymize_text(t) for t in tqdm(df['ocr_text'])]
```

```python
df.to_pickle('gdpr_clean_en_es.pkl')print('Anonymisation complete. File saved: gdpr_clean_en_es.pkl')
```
